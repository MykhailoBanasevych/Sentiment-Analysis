### Imports

```python
import os
import pandas as pd
import glob

```

### Path to all .csv files in a folder

```python
csv_files = glob.glob(r"C:\Users\HP\Documents\SentimentAnalysis\Sentiment-Analysis\Opinions\*.csv")
```

### Empty list

```python
dfs = []
for file in csv_files:
    try:
        df = pd.read_csv(file)
        dfs.append(df)
        print(f"Loaded {file} with shape {df.shape}")
    except Exception as e:
        print(f"Failed to read {file}: {e}")
```

### Loop and read of each file

```python
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)
```

### Combining into one data frame

```python
merged = pd.concat(dfs, ignore_index=True)
```

### Save to single file

```python
merged.to_csv("merged.csv", index=False)
```
